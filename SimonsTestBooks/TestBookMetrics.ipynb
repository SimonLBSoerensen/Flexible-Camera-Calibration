```python
num_samples = 500

# make a simple unit circle 
theta = np.linspace(0, 2*np.pi, num_samples)
a, b = 1 * np.cos(theta), 1 * np.sin(theta)

r = np.random.rand((num_samples))*10
rc = (r/10)*360
x, y = r * np.cos(theta), r * np.sin(theta)
```

```python
plt.figure(figsize=(5,5))

ax = plt.subplot(1,1,1)

points = np.array([x, y]).T
points_angels = np.array([np.rad2deg(theta)]).T
flexiblecc.Metrics.voronoi.plot_voronoi(points, points_angels, ax=ax)
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.savefig("voronoi.svg")
plt.show()
```
